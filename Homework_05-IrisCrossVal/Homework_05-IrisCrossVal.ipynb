# import modules

```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

# helper function(s)

# read iris data into Pandas data frame

```python
names = ["slength", "swidth", "plength", "pwidth", "class"]
data = pd.read_csv("iris.data", delimiter = "," , names = names)
print(data)
```

# plot data to investigate pattern(s)

```python
colors = {'Iris-setosa' : 'red', 'Iris-versicolor' : 'green', 'Iris-virginica':'purple'}
y = data['class']
c = [colors[x] for x in y]
pd.plotting.scatter_matrix(data, c=c)
```

# split data into feature matrix (X) and target labels (y)

```python
X = data.drop(['class'], axis = 1)
y = data['class']
```

# create decision tree classifier

```python
myDtree = tree.DecisionTreeClassifier()
```

# use K-fold cross validation to test classifier

```python
AccuracyVal = cross_val_score(myDtree, X, y, cv=5)
pred = cross_val_predict(myDtree, X, y, cv=5)
a = accuracy_score(y, pred)
p = precision_score(y, pred, average='micro')
r = recall_score(y, pred, average='micro')
```

# print results: accuracy, precision, and recall 

```python
print("accuracy: "+ str(a))
print("percision: "+ str(p))
print("recall: "+ str(r))
```

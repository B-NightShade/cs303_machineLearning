# import statements

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# python code to read census data into pandas data frame

```python
column_names = ["age", "workclass", "fnlwg", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"]
data = pd.read_csv('adult.data', header = None, delimiter = ', ', names = column_names)
```

# helper functions (if needed)

# python code to process, analyze, visualize trend #1

```python
hfincome =  np.where((data['sex']=='Female') & (data['income'] == '>50K'))

total = 0
values, counts =np.unique(hfincome, return_counts=True)
for count in counts:
    total += int(count)

lfincome =  np.where((data['sex']=='Female') & (data['income'] == '<=50K'))

total2 = 0
values, counts =np.unique(lfincome, return_counts=True)
for count in counts:
    total2 += int(count)
    

hmincome =  np.where((data['sex']=='Male') & (data['income'] == '>50K'))

total3 = 0
values, counts =np.unique(hmincome, return_counts=True)
for count in counts:
    total3 += int(count)

lmincome =  np.where((data['sex']=='Male') & (data['income'] == '<=50K'))

total4 = 0
values, counts =np.unique(lmincome, return_counts=True)
for count in counts:
    total4 += int(count)

men = data.loc[data['sex'] == 'Male']
women = data.loc[data['sex']=='Female']

a = np.shape(women)[0]
b = np.shape(men)[0]

perHf = (total / a)*100
perlf = (total2 / a)*100
perHm = (total3 / b)*100
perlm = (total4 / b) *100

plt.figure()
plt.title("sex vs income")
plt.xlabel("category of income")
plt.ylabel("percent in that group")
plt.bar('highFincome',perHf , color ='black', label = "fem high income")
plt.bar('highMincome',perHm , color ='red', label = "men high income")
plt.bar('lowFincome',perlf , color ='cyan', label = "fem low income")
plt.bar('lowMincome',perlm , color ='green', label = "men low income")
plt.legend()
plt.show()
```

# plaintext explanation of trend #1.  What did you find?  How does your graph show this trend?

When looking at this data both men and women have a higher percentage of people who fall in the lower income category(<=50). This can be seen by looking at the height of the bars where height correlates to the percent of people in a category. The taller the bar graph the higher the percent of men/females are in that group. Note that men and female are seperated so that females add up to 100% and men add up to 100%. Cyan and green are for the lower income families where cyan is female and green is men.
Men have a higher percentage of people who are in the higher income category while women hava a higher percentage of people in the lower income category.
In short the trend that you can see here is that a person is more likely to fall into the lower income category regardless of their sex. Both men and women find that higher percentages of their groups fall in the category where income is <=50K.

# python code to process, analyze, visualize trend #2

```python
white =  data['race']== 'White'
acp = data['race']== 'Asian-Pac-Islander'
aie = data['race']== 'Amer-Indian-Eskimo'
black = data['race'] == 'Black'
other = data['race'] == 'Other'

cgWhite = data['education-num'][white]
cgACP = data['education-num'][acp]
cgAIE = data['education-num'][aie]
cgBLACK = data['education-num'][black]
cgOTHER = data['education-num'][other]

plt.figure()
plt.title('education num and race')
plt.boxplot([cgWhite, cgACP, cgAIE, cgBLACK, cgOTHER])
labelg = ['white', 'Asian-Pac-Islander', 'Amer-Indian-Eskimo', 'black', 'other']
plt.xticks(ticks=[1,2,3,4,5], labels = labelg, rotation=45)
plt.xlabel("race")
plt.ylabel("education number")
plt.show()
```

# plaintext explanation of trend #2.  What did you find?  How does your graph show this trend?

The trend here comes in the varience and outliers of these different races. Those who identified as white and Asian-Pacfic-Islander have the largest varience in their education number. They also have fewer outliers in their data points. Amer-Indian-Eskimo have the lowest amount of varience and the most outliers to their data. Other fall somewhere between these groups in terms of varience and outliers.
Varience measures hwo far each value is from the mean of the data for that race. Varience is a way to show spread and is measured in boxplots by the height of the boxed in area on the illistration. The outliers play a part in the varience because they are data points significantly far from the center of data. Outliers appear as the dots outside the whiskers of the boxplot and they represent numbers that are more then 3 standard deviations from the mean and thus are not as likely to be seen when looking at a group of individuals in that category.
We can see that for those who are white and Asian-Pacific-Islander it is rare for them to get a low education number because they have outliers down near the one or two area. They also have the widest range of common education numbers in their group. Amer-Indian-Eskimo and black individuals tend to score somwhere in the middle at 8 to 10. straying from the middle of the education number is rare and makes you an outlier. They are very consistent with their scores being in the range of 8 to 10. The other group also tends to stick tot he center;however, they have enough low values that their outliers in the lower numbers of the eductaion number are restricted to 1 through 3.

# python code to process, analyze, visualize trend #3


```python
plt.figure()

plt.title("sex and education num")
men =  data['sex']== 'Male'
women = data['sex']== 'Female'

mED = data['education-num'][men]
wED = data['education-num'][women]

plt.hist(mED, alpha = 0.45, bins = 16, label ='men')
plt.hist(wED, alpha = 0.45, bins = 16, label ='women')
plt.ylabel("counts")
plt.xlabel("education number")
plt.legend()

plt.show()
```

# plaintext explanation of trend #3. What did you find? How does your graph show this trend?

This graph shows the trend that regardless of the sex of an individual of this census, they tend to land in the 8 to 12 range on the education number scale. histograms are good for showing the spread of data. We can see that the trend for overall education score is left skew(meaning that the higher counts tend to fall to the right of the graph and trail off to the left) in both of these groups. Since there is skew here, it shows that it is more likely that an individual at the time this census was taken had a higher education number. If the data displayed a more normal distribution, then they would tend to fall in the center of the histogram and if the data showed right skew, more people would have been given smaller education numbers on average. The histogram for women shows less spread into the higher numbers, but it is important to note here that the counts for the number of men and women in the census are not equal.
Men appear to have a bit more left skew then the women. This could be an affect of the difference in counts;however, if you look at the trend of how womens bars seem to rise and fall in the same bins as the men, this trend disappears near bin 15 and 16. This could mean that men have a slightly higher chance of hitting the education levels of 15 and 16 when this census was taken. This would need to be explored more heavily with statistics to see if it is a significant difference in those categories compared to the others.

# python code to process, analyze, visualize trend #4

```python
whiteH = np.where((data['income']== '>50K') & (data['race']=='White'))

WHtotal = 0
values, counts =np.unique(whiteH, return_counts=True)
for count in counts:
    WHtotal += int(count)

whiteL = np.where((data['income']== '<=50K') & (data['race']=='White'))

WLtotal = 0
values, counts =np.unique(whiteL, return_counts=True)
for count in counts:
    WLtotal += int(count)
    
apiH = np.where((data['income']== '>50K') & (data['race']=='Asian-Pac-Islander'))

APItotal = 0
values, counts =np.unique(apiH, return_counts=True)
for count in counts:
    APItotal += int(count)

apiL = np.where((data['income']== '<=50K') & (data['race']=='Asian-Pac-Islander'))

APILtotal = 0
values, counts =np.unique(apiL, return_counts=True)
for count in counts:
    APILtotal += int(count)
    
aieH = np.where((data['income']== '>50K') & (data['race']=='Amer-Indian-Eskimo'))

aietotal = 0
values, counts =np.unique(aieH, return_counts=True)
for count in counts:
    aietotal += int(count)

aieL = np.where((data['income']== '<=50K') & (data['race']=='Amer-Indian-Eskimo'))

aieLtotal = 0
values, counts =np.unique(aieL, return_counts=True)
for count in counts:
    aieLtotal += int(count)


bH = np.where((data['income']== '>50K') & (data['race']=='Black'))

btotal = 0
values, counts =np.unique(bH, return_counts=True)
for count in counts:
    btotal += int(count)

bL = np.where((data['income']== '<=50K') & (data['race']=='Black'))

bLtotal = 0
values, counts =np.unique(bL, return_counts=True)
for count in counts:
    bLtotal += int(count)
    
oH = np.where((data['income']== '>50K') & (data['race']=='Other'))

ototal = 0
values, counts =np.unique(oH, return_counts=True)
for count in counts:
    ototal += int(count)

oL = np.where((data['income']== '<=50K') & (data['race']=='Other'))

oLtotal = 0
values, counts =np.unique(oL, return_counts=True)
for count in counts:
    oLtotal += int(count)

whiteHI = WHtotal/(WHtotal+WLtotal)
whiteLI = WLtotal/(WHtotal+WLtotal)

apiHI = APItotal/(APItotal+APILtotal)
apiLI = APILtotal/(APItotal+APILtotal)

aieHI = aietotal/(aietotal+aieLtotal)
aieLI = aieLtotal/(aietotal+aieLtotal)

bHI = btotal/(btotal+bLtotal)
bLI = bLtotal/(btotal+bLtotal)

oHI = ototal/(ototal+oLtotal)
oLI = oLtotal/(ototal+oLtotal)

plt.figure()
plt.bar("white hi",whiteHI, color ='black', label = ">50K income")
plt.bar("white li",whiteLI, color ='red', label = "<=50K income")
plt.bar('Asian-Pac-Islander hi',apiHI, color ='black')
plt.bar("Asian-Pac-Islander li",apiLI, color ='red')
plt.bar('Amer-Indian-Eskimo hi',aieHI, color ='black')
plt.bar('Amer-Indian-Eskimo li',aieLI, color ='red')
plt.bar("black hi",bHI, color ='black')
plt.bar("black li",bLI, color ='red')
plt.bar("other hi",oHI, color ='black')
plt.bar("other li",oLI, color ='red')
plt.title("Race and income")
plt.ylabel("percent in category")
plt.xlabel("race category")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

# plaintext explanation of trend #4.  What did you find? How does your graph show this trend?

In this visualization we are looking at the percentage of each race that falls within the two categories of income. This visualization shows us that an overall trend for thois data is that a higher percentage of each race group is in the lower income bracket(<=50K the red bars). Those who fell int he race categories of white or Asian Pacific Islander had a higher percentage or chance to fall within the higher income group(>50K black bars). This graph shows this trend with the hieghts of the bars. The heights correspond ot the percentage of that group that falls within a particular category. By making the percentages group based instead of basing them on everyone in the data set allows us to abstract away from the fact that some groups have a lot more data points to look at. Instead we are looking at a ratio for each group and can compare across groups without having to worry as much about the count differences across groups.
